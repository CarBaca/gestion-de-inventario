##### El dashboard se divide de la siguiente manera:
##### Portada 

##### En la portada nos vamos a encontrar con cinco botones que nos ayudan a navegar por las pginas del tablero.


##### Compras 

##### En esta pgina, el usuario puede consultar de un vistazo una descripcin general completa de la informacin ms relevante sobre las compras. Los datos se pueden filtrar por categorias, productos, meses y ciudades, lo que proporciona una visualizacin personalizada de como se distribuyen las compras.
* **Las tarjetas superiores resaltan los indicadores claves de las compras, proporcionando una descripcin rpida de la situacin.**
* **El grfico de lneas muestra como se desenvuelven las ventas a los largo del tiempo .**
* **Un segundo grfico de columnas que proporciona informacin sobre los productos que mas se compra .**
* **A la izquierda del tablero se encuentra los botones para navegar entre las pginas.**
* **El ltimo grfico de donut chart (grfico de rosquilla) que muestra la distribucin proporcional de las compras entre estas dos categoras.**

##### Siguiendo con el anlisis de compras en una segunda seccin tenemos las tendencias:

* **El grfico de reas apiladas explica las tendencias de compras por categora para dos tipos de productos: licores y vinos. Es relevante porque permite visualizar cmo varan las compras de estos productos a lo largo del tiempo, en este caso, entre diciembre de 2015 y febrero de 2016.**
* **El segundo grfico de barras  ilustra las cantidades de compras segn diferentes tamaos de presentacin.
Los tamaos incluyen 750 ml, 1L, 1.75L.**


#####  Ventas 

##### En esta seccin profundizamos en los datos especficos de las ventas, proporcionando una descripcin ms detallada de la situacin.
* **En la parte superior encontramos con los filtros por categora, por productos donde se puede escribir y buscar el producto interesado, tambin se puede filtrar meses y ciudades.**
* **Abajo de los filtros se encuentran los KPIS  mas importantes.**
* **Un tacometro de tiempo de ciclo de pedidos que es crucial para evaluar la eficiencia del proceso de adquisicin y entrega de productos.**
* **Un grfico de torta que proporciona una visin rpida de cmo se comparan las ventas entre estas dos categoras.**
* **Como tercer grfico, encontramos el de reas que muestra la distribucin de las ventas en el periodo seleccionado.**
* **A la izquierda del tablero se encuentra los botones para navegar entre las pginas.**
* **En el margen derecho se puede observar marcadores con los nombres de los producto mas vendidos,donde se puede hacer clic y la imagen del producto va cambiando.**

##### Transacciones

##### Esta pgina proporciona una descripcin detallada de los movimientos de la empresa.

* **Al igual de las pginas enteriores en la parte superior podemos encontrar los filtros.**
* **A la izquierda del tablero se eencuentra los botones para navegarentre las pginas.**
* **En la seccin grficos podemos ver un grfico de reas, donde se puede observar como las cantidades fluctuan a lo largo del tiempo.**
* **Un segundo grfico de barras que muestra las medidas que mas se venden.**
* **Por ltimo una tabla que describe cada producto, las unidades vendidas y el precio.**
* **En la seccin de "esquema" podemos observar un esquema jerarquico, que va mostrando las ventas segn la clasificacin y los productos.**

##### Stock
##### La pgina de "Stock" proporciona una visin integral del estado actual del inventario.


* **En la parte superior se traen los mismo filtros que las pginas anteriores.**
* **Luego se encuentran los tacometros como principales KPIS: Tasa de venta de directa ( expresa el porcentaje de inventario vendido en comparacin con la cantidad de inventario recibido del fabricante o proveedor ), Rotacin de Inventario (Mide la eficiencia con la que se vende el inventario), Ciclo de conversin de efectivo (Es el tiempo en que el dinero vuelve a ser dinero ).**
* **A la izquierda del tablero se encuentra los botones para navegar entre las pginas.**
* **Por ltimo una matriz, que describe en que estado se encuentran cada unos de los productos.**


##### Rentabilidad 
##### La pgina de "Rentabilidad" ofrece una visin detallada del desempeo financiero de la empresa.


* **En la parte superior se pueden aplicar filtros por categoras, productos, meses y ciudades.**
* **Encontramos tarjetas KPIS, indicaroes de rentabilidad: margen bruto y valor del stock.**
* **Un grfico de columnas agrupadas y de lneas que muestra en que nivel esta el stock en comparacion con el stock de seguridad y el stock mximo.**
* **A la izquierda del tablero se encuentra los botones para navegar entre las pginas.**
* **Finalmente un mapa de calor que indica cuales son los productos mas rentables ( top 10 productos).**
